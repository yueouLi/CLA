```python
#@title Solution

# To implement the decision rule for the prediction, 
# we can compute the dot product of the feature vector 
# counts and the weight vector self.weights, and check 
#if the result is greater than zero. If the result is 
# greater than zero, we predict that the instance belongs
# to class 1, otherwise we predict that it belongs to 
# class -1.

# Here is the code for the prediction(self, counts) 
# function:
```

```python
#@title Solution

import numpy as np

def prediction(self, counts):
    # Compute the dot product of the feature vector and the weight vector
    dot_product = np.dot(counts, self.weights)
    # Check if the result is greater than zero
    if dot_product > 0:
        return 1
    else:
        return -1

# To implement the update for the instance, we need
# to adjust the weights based on the prediction error.
# If the instance is misclassified, we need to add the
# feature vector to the weight vector if the instance
# belongs to class 1, or subtract it if the instance
# belongs to class -1. If the instance is correctly 
# classified, we don't need to do anything.

def update(self, instance):
    # Get the label and feature vector of the instance
    label = instance[0]
    counts = instance[1:]
    # Make a prediction for the instance
    prediction = self.prediction(counts)
    # Check if the prediction is correct
    if prediction != label:
        # Adjust the weights based on the prediction error
        if label == 1:
            self.weights += counts
        else:
            self.weights -= counts

# Note that we assume that the counts vector contains 
# the feature counts for the instance, and that the 
# label is the first element of the instance. We also 
# assume that self.weights is a vector of the same length
# as counts that contains the current weight values.
```

* Ein Trainingszyklus auf den Trainingsdaten.
* Für jede Instanz wird Update ausgeführt (falls Fehlklassifikation).
* Vor jeder Iteration sollten die Daten neu gemischt werden.

```python
def training_iteration(self, dataset):
  dataset.shuffle()
  for instance in dataset.instance_list:
    self.update(instance)
```
